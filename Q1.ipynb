```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
print("First 10 rows of the breast cancer dataset:")
print(df.head(10))


print("\nBasic statistics for each feature:")
print(df.describe())


print("\nChecking for missing values:")
print(df.isnull().sum())
print("\nHandling missing values is important because missing data can lead to inaccurate model predictions and biased results.")

```
